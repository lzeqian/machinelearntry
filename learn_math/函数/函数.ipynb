```python
import matplotlib
import numpy as np;
import matplotlib.pyplot as plot;
"""
常熟函数 y=1
"""
def const():
    x=np.random.uniform(1,10,(1,10))[0]
    y=np.full((1,10),1)[0];
    plot.plot(x,y,'-')
    font = matplotlib.font_manager.FontProperties(fname="simsun.ttc") 
    plot.title("常值函数", fontproperties=font)
    plot.show()
const()
```

```python
import numpy as np;
import matplotlib.pyplot as plot;
from commons.common import setXY
"""
sgn函数
即 x>0,sgnx= 1
   x=0,sgnx= 0
   x<0,sgnx=-1
"""
def sgn():
    setXY()
    x=np.random.uniform(-10,10,(1,1000))[0]
    y=[1 if v>0 else (0 if v==0 else -1)  for v in x]
    font = matplotlib.font_manager.FontProperties(fname="simsun.ttc") 
    plot.title("分段常值函数", fontproperties=font)
    plot.plot(x,y,'.')
    plot.show()
sgn()
```

```python
import numpy as np;
import matplotlib.pyplot as plot;
from commons.common import setXY
%matplotlib inline
"""
幂函数
即 y=x
   y=x^2
   y=x^3
"""
def sgn():
    x=np.linspace(-10,10,1000)
    y=x;
    y1=x*x;
    y2=x*x*x;
    x1=np.linspace(-100,100,1000)
    y3=1.0/x1*1.0
    plot.figure(figsize=(8,8), dpi=80)
    plot.figure(1)
    plot.subplot(221)
    font = matplotlib.font_manager.FontProperties(fname="simsun.ttc") 
    plot.title("直线函数", fontproperties=font)
    plot.plot(x,y)
    setXY()
    plot.subplot(222)
    plot.title("a=2幂函数", fontproperties=font)
    setXY()
    plot.plot(x,y1)
    #把画板分成两行一列取第二行
    plot.subplot(223) 
    setXY()
    plot.title("a=3幂函数", fontproperties=font)
    plot.plot(x,y2,'-r')
    plot.subplot(224) 
    setXY()
    plot.title("a=-1幂函数", fontproperties=font)
    plot.plot(x1,y3,'-r')
    plot.show()
sgn()
```

```python
import numpy as np;
import matplotlib.pyplot as plot;
from commons.common import setXY
%matplotlib inline
"""
绝对值函数
即 y=|x|
   y=-x x<=0
   y=x x>0
   
"""
def abs():
    x=np.linspace(-10,10,1000)
    y=np.abs(x);
    setXY()
    font = matplotlib.font_manager.FontProperties(fname="simsun.ttc") 
    plot.title("绝对值函数", fontproperties=font)
    plot.plot(x,y)
    plot.show()
abs()
```

```python
import numpy as np;
import matplotlib.pyplot as plot;
from commons.common import setXY
%matplotlib inline
"""
取整函数，函数y=[x]称为取整函数，也称高斯函数。其中不超过实数x的最大整数称为x的整数部分，记作[x]。该函数被广泛应用于数论，函数绘图和计算机领域。
即 y=[x]
   y=-x x<=0
   y=x x>0
   
"""
def int():
    x=3.5
    print(np.floor(x))
    x=np.linspace(-10,10,1000)
    y=np.floor(x);
    setXY()
    font = matplotlib.font_manager.FontProperties(fname="simsun.ttc") 
    plot.title("取整函数", fontproperties=font)
    plot.plot(x,y,'.')
    plot.show()
int()
```

```python
import matplotlib.pyplot as plot;
import matplotlib;
from commons.common import setXY
import numpy as np
%matplotlib inline
#log函数
def log():
    x=3.5
    print(np.log2(x))
    x=np.linspace(-10,10,1000)
    y=np.log2(x);
    #这里就是log已0.5为底数 i为x的对数函数
    #底数是0-1之间是单调递减 大于1是单调递增
    y1=[np.log(i)/np.log(0.5)for i in x]
    setXY()
    font = matplotlib.font_manager.FontProperties(fname="simsun.ttc") 
    plot.title("log函数", fontproperties=font)
    plot.plot(x,y)
    plot.plot(x,y1)
    plot.show()
log()
```
