```python
import numpy as np
import matplotlib.pyplot as plt

# 生成一组随机样本数据
mu = 0  # 均值
sigma = 5  # 标准差
samples = np.random.normal(mu, sigma, 1000)

# 绘制密度曲线
plt.hist(samples, bins=30, density=True, alpha=0.7, color='c')

# 计算密度曲线的x和y值
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
y = 1/(sigma * np.sqrt(2*np.pi)) * np.exp(-(x - mu)**2 / (2*sigma**2))

# 绘制密度曲线
plt.plot(x, y, color='r')

# 添加标题和轴标签
plt.title('Density Curve: Mean=0, Standard Deviation=2')
plt.xlabel('Value')
plt.ylabel('Density')

# 显示图形
plt.show()
```
